#Bar charts and Histograms

```
import matplotlib.pyplot as plt

days = [1,2,3,4,5]
sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]


plt.plot([],[], color= 'r', label = 'sleeping', linewidth=8)
plt.plot([],[], color= 'g', label = 'eating', linewidth=8)
plt.plot([],[], color= 'b', label = 'working', linewidth=8)
plt.plot([],[], color= 'm', label = 'playing', linewidth=8)


plt.stackplot(days, sleeping, eating, working, playing,
              colors=['r','g','b','m'])
              

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Interesting Graph\nCheck it out')
```
